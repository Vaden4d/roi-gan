```python
import sys
import matplotlib.pyplot as plt

sys.path.append('..')
%load_ext autoreload
%autoreload 2
from utils.rois import squared_roi, gaussian_roi, mixture_roi
```

```python
def plot_images(function, nrows=3, ncols=8):

    # plot the rois in the grid
    fig, axes = plt.subplots(nrows, ncols, figsize=(15, 5), dpi=100)

    for i in range(nrows):
        for j in range(ncols):
            mask = function()
            axes[i, j].imshow(mask, cmap='gray')
            axes[i, j].axis('off')
```

```python
plot_images(squared_roi)
```

```python
plot_images(gaussian_roi)
```

```python
plot_images(mixture_roi)
```
